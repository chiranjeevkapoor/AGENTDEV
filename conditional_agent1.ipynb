CONDITIONAL AGENT EXERCISE 

```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class AgentState(TypedDict):
    number1:int
    number2:int
    operation1:str
    number3:int
    number4:int
    operation2:str
    finalanswer1:int
    finalanswer2:int


def adder_node1(state:AgentState)->AgentState:
    state["finalanswer1"] = state["number1"] + state["number2"]
    return state

def sub_node1(state:AgentState)->AgentState:
    state["finalanswer1"] = state["number1"] - state["number2"]
    return state

def adder_node2(state:AgentState)->AgentState:
    state["finalanswer2"] = state["number3"] + state["number4"]
    return state

def sub_node2(state:AgentState)->AgentState:
    state["finalanswer2"] = state["number3"] - state["number4"]
    return state


def decide_node1(state:AgentState)->AgentState:
    if state["operation1"]=="+":
        return "addition1"
    elif state["operation1"]=="-":
        return "subtraction1"

def decide_node2(state:AgentState)->AgentState:
    if state["operation2"]=="+":
        return "addition2"
    elif state["operation2"]=="-":
        return "subtraction2"
```

```python
graph = StateGraph(AgentState)

graph.add_node("add1", adder_node1)
graph.add_node("add2", adder_node2)
graph.add_node("sub1", sub_node1)
graph.add_node("sub2", sub_node2)
graph.add_node("router1",lambda state:state)
graph.add_node("router2", lambda state:state)
graph.add_edge(START, "router1")

graph.add_conditional_edges(
    "router1",
    decide_node1,
    {
        "addition1":"add1",
        "subtraction1":"sub1"
    }
)
graph.add_edge("add1","router2")
graph.add_edge("sub1","router2")
graph.add_conditional_edges(
    "router2",
    decide_node2,
    {
        "addition2":"add2",
        "subtraction2":"sub2"
    }
)

graph.add_edge("add2",END)
graph.add_edge("sub2",END)
```

```python
app = graph.compile()
```

```python
answer = app.invoke({"number1":2,"number2":4,"number3":7,"number4":9,"operation1":"+","operation2":"-"})
```
