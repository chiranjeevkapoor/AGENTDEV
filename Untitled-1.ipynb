LOOP EXERCISE

```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, List
import random

class AgentState(TypedDict):
    name:str
    guesses:List[int]
    upper_bound:int
    lower_bound:int
    attempts:int
    target:int
    hint:str
    answer:str

def setup_node(state:AgentState)->AgentState:
    """This agent is for setup"""
    state["attempts"] = 0
    state['guesses'] = []
    state['hint'] = f"hi, {state['name']},We are playing a guessing game, we will guess a number b/w {state['lower_bound']} and {state['upper_bound']}, we have {state['attempts']} attempts."
    state['target'] = random.randint(state['lower_bound'], state['upper_bound'])
    print(state['hint'])
    return state

def guess_node(state:AgentState)->AgentState:
    """This node guesses the number"""
    possible_guesses = [i for i in range(state['lower_bound'], state['upper_bound']+1) if i not in state['guesses']]
    if possible_guesses:
        guess = random.choice(possible_guesses)
    else:
        guess = random.randint(state['lower_bound'], state['upper_bound'])
    
    state['guesses'].append(guess)
    state['attempts']+=1
    print(f"Attempt {state['attempts']}: Guessing {guess} (Current range: {state['lower_bound']}-{state['upper_bound']})")
    return state


def hint_node(state:AgentState):
    """This node compares"""
    guessed_number = state['guesses'][-1]

    if guessed_number<state['target']:
        state['lower_bound'] = guessed_number+1
        return "guess"
    elif guessed_number>state['target']:
        state['upper_bound'] = guessed_number-1
        return "guess"
    else:
        print(f"You guessed it correctly ${state['target']} == ${guessed_number} in {state['attempts']} attempts.")
        return "EXIT"
```

```python
graph = StateGraph(AgentState)
graph.add_node("setup", setup_node)
graph.add_node("guess", guess_node)
graph.add_node("hint", lambda state:state)

graph.add_edge("setup", "guess")
graph.add_edge("guess", "hint")
graph.add_conditional_edges(
    "hint",
    hint_node,
    {
        "guess":"guess",
        "EXIT":END
    }
)
graph.set_entry_point("setup")
```

```python
app = graph.compile()
```

```python
answer = app.invoke({'name':'bob', 'lower_bound':1,'upper_bound':20 })
print(answer)
```
