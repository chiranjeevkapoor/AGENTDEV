LOOPING GRAPH

```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, List
import random
```

```python
class AgentState(TypedDict):
    name:str
    numbers:List[int]
    counter:int

def greeting_node(state:AgentState)->AgentState:
    """This node adds a greeting."""
    state['name'] = f"Hi there, {state['name']}"
    state['counter'] = 0
    return state

def random_node(state:AgentState)->AgentState:
    """This node adds a random number to the list"""
    state['numbers'].append(random.randint(0,10))
    state['counter'] +=1
    return state

def should_continue(state: AgentState)->AgentState:
    """This node decides if a loop should continue or not."""
    if state['counter']<5:
        print("ENTERING LOOP")
        return "loop"
    else:
        return "EXIT"
    
graph = StateGraph(AgentState)
```

```python
graph.add_node("greetings",greeting_node)
graph.add_node("random", random_node)

graph.add_edge("greetings", "random")

graph.add_conditional_edges(
    "random",
    should_continue,
    {
        "loop":"random",
        "EXIT": END
    }
)
graph.set_entry_point("greetings")
```

```python
app = graph.compile()

answer = app.invoke({"name":"bob",'counter':-1231, 'numbers':[]})
print(answer)
```
